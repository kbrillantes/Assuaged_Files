```python
import tweepy
import time
import pandas as pd
import json
import csv
import preprocessor as p
import numpy as np
```

```python
consumer_key=''
consumer_secret=''
access_token_secret=''
access_token=''
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
```

```python
txts = []

for tweet in tweepy.Cursor(api.search,
            q='@forksoverknives -filter:retweets',
            lang="en").items(5000):
    txts.append(json.loads(json.dumps(tweet._json)))
    
len(txts)
```

```python
txts[0]
```

```python
time = []
twtid = []
text = []
favorite_count = []
retweet_count=[]
hashtags=[]
favorite_count=[]
screen_name=[]

for line in txts:
    time.append(line['created_at'])
    twtid.append(line['id'])
    screen_name.append(line['user']['screen_name'])
    text.append(line['text'])
    favorite_count.append(line['favorite_count'])
    retweet_count.append(line['retweet_count'])
    hashtags.append(line['entities']['hashtags'])
```

```python
p.clean(text[1])
text_clean = []
for line in text:
    text_clean.append(p.clean(line))
text_clean[:1]
# create data frame
df = pd.DataFrame(list(zip(time, twtid, screen_name, text_clean, favorite_count, retweet_count, hashtags)),
                             columns=['created_at', 'tweet_id','screen_name','text','favorite_count','retweet_count',
                                     'hashtags'])
df
```

```python
import re
import string

#remove '[]' in hashtags
clean_df.hashtags = clean_df.hashtags.replace('[]','')
#convert hashtags into string
clean_df['hashtags']=clean_df['hashtags'].apply(str)
# look behind
re.findall(r"(?<='text':\s\')\w+",clean_df.hashtags[3])
' '.join(re.findall(r"(?<='text':\s\')\w+",clean_df.hashtags[3]))
# apply to every rows
clean_df.hashtags=clean_df.hashtags.apply(lambda row: ' '.join(re.findall(r"(?<='text':\s\')\w+",row)))
#check again
clean_df.hashtags.value_counts()
```

```python
# Convert timestamp datatype to datetime
clean_df['created_at'] = pd.to_datetime(clean_df['created_at'], infer_datetime_format=True)
#Check again
clean_df.info()
```
